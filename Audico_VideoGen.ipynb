## **1. Call the Text to video pipeline with default video length and chunk size**

```python
import torch
import imageio
from diffusers import TextToVideoZeroPipeline
import numpy as np

# pretrained stable diffusion model to execute the text to Video data
model_id = "runwayml/stable-diffusion-v1-5"
pipe = TextToVideoZeroPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")
```

### 1.1 Example 1 - Generate a small video for the prompts

```python
prompt = "A panda is playing guitar on times square"
result = pipe(prompt=prompt).images
result = [(r * 255).astype("uint8") for r in result]
imageio.mimsave("video.mp4", result, fps=4)
```

## Generate the Video pipeline based on the video length and chunk size params

```python
seed = 0

# Provide the three parameter as input
# Prompt - Text to  generate the video
# video_length - decide the length of the video, can able to increase the length of video by adjusting the length
# chunk_size - process the vidoe in chunks, suitable to bigger senterces
def generateVideo(prompt, video_length, chunk_size):
  # Generate the video chunk-by-chunk
  result = []
  chunk_ids = np.arange(0, video_length, chunk_size - 1)
  generator = torch.Generator(device="cuda")
  for i in range(len(chunk_ids)):
      print(f"Processing chunk {i + 1} / {len(chunk_ids)}")
      ch_start = chunk_ids[i]
      ch_end = video_length if i == len(chunk_ids) - 1 else chunk_ids[i + 1]
      # Attach the first frame for Cross Frame Attention
      frame_ids = [0] + list(range(ch_start, ch_end))
      # Fix the seed for the temporal consistency
      generator.manual_seed(seed)
      output = pipe(prompt=prompt, video_length=len(frame_ids), generator=generator, frame_ids=frame_ids)
      result.append(output.images[1:])

  # Concatenate chunks and save
  result = np.concatenate(result)
  result = [(r * 255).astype("uint8") for r in result]
  return result
```

Process first 15 sentence with video length of 150 and chunk size of 15.

```python
# process first 15 sentence from JSOn stories files, by adjusting the lower and upper index, it is possible to change the sentence input
upperIndex = 15
lowerIndex = 0
video_length = 150
chunk_size = 15
promptIds = np.arange(lowerIndex, upperIndex)
promptText = ''

for i in range(len(promptIds)):
  promptText += ls[promptIds[i]] + '.'

result = generateVideo(promptText, video_length, chunk_size)
save_path2 = '/content/drive/MyDrive/video_test.mp4'
imageio.mimsave(save_path2, result, fps=4)
#print(promptIds[i])
```
